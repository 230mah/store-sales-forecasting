```python
# Importar bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar dados de treino com o caminho correto
train = pd.read_csv('../data/raw/train.csv', parse_dates=['date'])

# Mostrar as primeiras linhas
train.head()
```

# Exploração Inicial dos Dados de Vendas

Neste notebook, começamos carregando o dataset principal de treino contendo as vendas diárias em várias lojas e categorias.

As colunas principais são:
- `date`: data da venda
- `store_nbr`: número da loja
- `family`: categoria do produto
- `sales`: quantidade vendida
- `onpromotion`: se o produto estava em promoção

Esse conjunto de dados será a base para todas as análises e previsões futuras.


```python
# Informações gerais do dataset
train.info()
```

# Informações Gerais

Aqui verificamos o tipo das colunas, quantidade de registros e presença de valores nulos. Isso é fundamental para entender a qualidade dos dados antes de avançar.


```python
# Estatísticas descritivas das vendas
train['sales'].describe()
```

# Estatísticas Básicas de Vendas

Visualizamos medidas como média, mediana, máximo e mínimo das vendas para entender a distribuição dos dados.


```python
# Vendas totais por loja
sales_by_store = train.groupby('store_nbr')['sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=sales_by_store.index, y=sales_by_store.values, palette='viridis')
plt.title('Vendas Totais por Loja')
plt.xlabel('Número da Loja')
plt.ylabel('Total de Vendas')
plt.grid(axis='y')
plt.show()
```

# Vendas Totais por Loja

Este gráfico mostra o total de vendas acumuladas em cada loja. Podemos identificar quais lojas possuem maior volume de vendas e entender a concentração das receitas.


```python
# Vendas totais por categoria (family)
sales_by_family = train.groupby('family')['sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(14,6))
sns.barplot(x=sales_by_family.index, y=sales_by_family.values, palette='magma')
plt.title('Vendas Totais por Categoria de Produto')
plt.xlabel('Categoria (Family)')
plt.ylabel('Total de Vendas')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

# Vendas Totais por Categoria de Produto

Este gráfico mostra o total de vendas acumuladas para cada categoria de produto.

Isso ajuda a identificar quais famílias de produtos têm maior impacto nas vendas e pode direcionar decisões estratégicas como estoque e promoções.


```python
# Análise temporal: vendas diárias totais ao longo do tempo
daily_sales = train.groupby('date')['sales'].sum()

plt.figure(figsize=(15,5))
daily_sales.plot()
plt.title('Vendas Diárias Totais ao Longo do Tempo')
plt.xlabel('Data')
plt.ylabel('Vendas Totais')
plt.grid(True)
plt.show()
```

# Vendas Diárias Totais ao Longo do Tempo

Este gráfico mostra o comportamento das vendas totais em todas as lojas ao longo do tempo.

É possível observar tendências sazonais, picos e possíveis efeitos de eventos, o que é fundamental para o planejamento.


```python
# Análise das vendas médias por loja
avg_sales_by_store = train.groupby('store_nbr')['sales'].mean().sort_values(ascending=False)

plt.figure(figsize=(14,6))
sns.barplot(x=avg_sales_by_store.index.astype(str), y=avg_sales_by_store.values, palette='coolwarm')
plt.title('Média de Vendas por Loja')
plt.xlabel('Número da Loja')
plt.ylabel('Média de Vendas')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

# Média de Vendas por Loja

Este gráfico apresenta a média de vendas por loja.

Ele é útil para identificar quais lojas possuem melhor desempenho em média, possibilitando focar análises e estratégias específicas para as lojas com menor performance.


```python
# Verificar a relação entre vendas e promoções
promo_sales = train.groupby('onpromotion')['sales'].mean()

plt.figure(figsize=(8,5))
sns.barplot(x=promo_sales.index.astype(str), y=promo_sales.values, palette='Set2')
plt.title('Impacto das Promoções nas Vendas Médias')
plt.xlabel('Promoção (0 = Não, 1 = Sim)')
plt.ylabel('Média de Vendas')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

# Impacto das Promoções nas Vendas Médias

Aqui analisamos a média de vendas quando o produto está em promoção versus quando não está.

Observa-se se promoções influenciam positivamente as vendas, informação vital para estratégias de marketing e precificação.


```python
# Carregar dados de feriados
holidays = pd.read_csv('../data/raw/holidays_events.csv', parse_dates=['date'])

# Filtrar apenas feriados que não são transferidos
holidays = holidays[(holidays['transferred'] == False) & (holidays['type'] == 'Holiday')]

# Mostrar primeiros registros para conferência
holidays.head()
```

# Dados de Feriados

Este dataset contém informações sobre os feriados que podem impactar as vendas.

Selecionamos apenas os feriados oficiais, excluindo os transferidos para outro dia e eventos que não são feriados, para focar em impactos diretos.


```python
# Visualizar a quantidade de feriados por mês
holidays['month'] = holidays['date'].dt.month

plt.figure(figsize=(10,5))
sns.countplot(x='month', data=holidays, palette='magma')
plt.title('Número de Feriados por Mês')
plt.xlabel('Mês')
plt.ylabel('Quantidade de Feriados')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

# Distribuição dos Feriados ao Longo do Ano

Este gráfico mostra em quais meses do ano os feriados acontecem com maior frequência.

Esse padrão ajuda a entender possíveis picos ou quedas sazonais nas vendas relacionadas a feriados.


```python
# Analisar vendas médias em dias de feriado vs dias normais

# Criar coluna indicando se a data é feriado
train['is_holiday'] = train['date'].isin(holidays['date'])

holiday_sales = train.groupby('is_holiday')['sales'].mean()

plt.figure(figsize=(6,4))
sns.barplot(x=holiday_sales.index.astype(str), y=holiday_sales.values, palette='pastel')
plt.title('Média de Vendas em Dias de Feriado vs Dias Normais')
plt.xlabel('Dia de Feriado')
plt.ylabel('Média de Vendas')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

# Vendas Médias em Dias de Feriado vs Dias Normais

Aqui avaliamos o efeito dos feriados nas vendas médias.

Comparando as médias, podemos inferir se os feriados elevam ou reduzem as vendas, importante para planejamento operacional e de estoque.


```python
# Extrair dia da semana e mês da data
train['day_of_week'] = train['date'].dt.day_name()
train['month'] = train['date'].dt.month_name()

# Média de vendas por dia da semana
dow_sales = train.groupby('day_of_week')['sales'].mean()

# Para ordenar os dias corretamente
dias_semana = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
dow_sales = dow_sales.reindex(dias_semana)

plt.figure(figsize=(10,5))
sns.barplot(x=dow_sales.index, y=dow_sales.values, palette='coolwarm')
plt.title('Média de Vendas por Dia da Semana')
plt.xlabel('Dia da Semana')
plt.ylabel('Média de Vendas')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

# Média de Vendas por Dia da Semana

O gráfico mostra como as vendas variam ao longo dos dias da semana.

É possível observar padrões, como maior movimento em determinados dias, o que pode ajudar em estratégias de marketing e logística.


```python
# Média de vendas por mês
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 
               'July', 'August', 'September', 'October', 'November', 'December']

monthly_sales = train.groupby('month')['sales'].mean()
monthly_sales = monthly_sales.reindex(month_order)

plt.figure(figsize=(12,5))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values, marker='o', color='darkblue')
plt.title('Média de Vendas por Mês')
plt.xlabel('Mês')
plt.ylabel('Média de Vendas')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Média de Vendas por Mês

```python
# Média de vendas por loja
store_sales = train.groupby('store_nbr')['sales'].mean().sort_values(ascending=False)

plt.figure(figsize=(14,6))
sns.barplot(x=store_sales.index, y=store_sales.values, palette='viridis')
plt.title('Média de Vendas por Loja')
plt.xlabel('Número da Loja')
plt.ylabel('Média de Vendas')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

Este gráfico mostra o total de vendas diárias somadas para todas as lojas e categorias.

- **Eixo X:** Data (período da base de dados).  
- **Eixo Y:** Total de vendas do dia.

### Observações:

- É possível observar tendências de crescimento ou queda nas vendas ao longo do tempo.  
- Sazonalidades e picos podem indicar períodos de alta demanda, como feriados ou promoções.

```python
import seaborn as sns

# Configura o estilo do seaborn para gráficos mais bonitos
sns.set(style="whitegrid")

# Cria dataframe para facilitar o seaborn trabalhar
promo_df = train.groupby('onpromotion')['sales'].mean().reset_index()
promo_df['onpromotion'] = promo_df['onpromotion'].map({0: 'Sem Promoção', 1: 'Com Promoção'})

plt.figure(figsize=(7,5))
ax = sns.barplot(x='onpromotion', y='sales', data=promo_df, palette=['#d62728', '#2ca02c'])

ax.set_title('Vendas Médias: Dias com e sem Promoção', fontsize=14)
ax.set_xlabel('')
ax.set_ylabel('Vendas Médias')
plt.ylim(0, promo_df['sales'].max() * 1.1)  # Espaço extra no topo do gráfico

# Mostrar valores exatos no topo das barras
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height:.2f}', (p.get_x() + p.get_width() / 2, height),
                ha='center', va='bottom', fontsize=11)

plt.tight_layout()
plt.show()
```

Essa análise fornece uma visão geral da contribuição de cada loja para o resultado global.


```python
# Calcular vendas médias diárias por loja
daily_avg_sales_store = train.groupby('store_nbr').apply(lambda x: x['sales'].sum() / x['date'].nunique()).sort_values(ascending=False)

# Plotar
plt.figure(figsize=(12,6))
daily_avg_sales_store.plot(kind='bar', color='cornflowerblue')
plt.title('Vendas Médias Diárias por Loja')
plt.xlabel('Número da Loja')
plt.ylabel('Média Diária de Vendas')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Este gráfico mostra a média diária de vendas para cada loja durante o período analisado.

- É útil para comparar lojas independentemente do tempo ou número de registros.
- Identifica quais lojas têm consistência nas vendas, podendo indicar melhor operação ou localização.
